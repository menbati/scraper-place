# Elastic search statistics

Make queries to elasticsearch using the python API.

```python
from elasticsearch import Elasticsearch
from matplotlib import pyplot as plt
```

```python
client = Elasticsearch()
```

```python
response = client.search(
    index="dce",
    body={
        "_source": False,
        'size' : 10000,
        'query': {
            'match_all' : {},
        },
    },
)
```

```python
hits = response['hits']['hits']
len(hits)
```

```python
ids = [hit['_id'] for hit in hits]
```

```python
content_lengths = []

for i, _id in enumerate(ids):
    response = client.get(index='dce', doc_type='dce', id=_id)
    content_lengths.append(len(response['_source']['content']))
    if (i % 100) == 0:
        print(i)
```

```python
content_lengths
```

```python
plt.hist(content_lengths, bins=100)
```

```python
plt.hist(
    [
        cl
        for cl in content_lengths
        if cl > 2000000
    ],
    bins=100
)
```

```python
[
    cl
    for cl in content_lengths
    if cl > 10000000
]
```

From these results, let's say the content should not exceed 10Mb.

If it does, only the first 5Mb and the last 5Mb are kept.

```python
big_dce_list = [
    (_id, cl)
    for _id, cl in zip(ids, content_lengths)
    if cl > 10000000
]
big_dce_list
```
