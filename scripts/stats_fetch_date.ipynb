```python
import json
import collections

from dateutil import parser
from matplotlib import pyplot as plt
```

```python
with open('fetch_datetime_export.json', 'r') as f:
    fetch_datetime_data = json.load(f)
```

```python
len(fetch_datetime_data)
```

```python
fetch_datetime_data[0]['fetch_datetime']
```

```python
datetimes = [
    parser.parse(e['fetch_datetime'])
    for e in fetch_datetime_data
]
```

```python
datetimes
```

```python
plt.figure(figsize=(15, 6), dpi=80)
plt.hist(datetimes, bins=300)
```

```python
days = [
    (dt.year, dt.month, dt.day)
    for dt in datetimes
]
```

```python
print(sorted(list(collections.Counter(days).items())))
```
